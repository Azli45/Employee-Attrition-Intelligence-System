## Observation:

Dataset is imbalanced.

Employee attrition cases are significantly lower
than non-attrition cases.

This imbalance must be considered
during model training.

```python
# Target visualization

plt.figure(figsize=(5,4))

sns.countplot(
data=df,
x='Attrition'
)

plt.title('Employee Attrition Distribution')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
data=df,
x='Department',
hue='Attrition'
)

plt.xticks(rotation=15)

plt.title(
'Department Wise Attrition'
)

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
data=df,
x='OverTime',
hue='Attrition'
)

plt.title(
'Overtime Impact on Attrition'
)

plt.show()
```

 Check whether employees doing overtime
leave more frequently.

```python
plt.figure(figsize=(8,5))

sns.histplot(
df['MonthlyIncome'],
bins=30,
kde=True
)

plt.title(
'Monthly Income Distribution'
)

plt.show()
```

Salary spread
Outliers
Skewness

```python
plt.figure(figsize=(8,5))

sns.boxplot(
data=df,
x='Attrition',
y='Age'
)

plt.title(
'Age vs Attrition'
)

plt.show()
```

## Correlation Heatmap

Numeric columns:

```python
numeric_df=df.select_dtypes(
include=np.number
)
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(

numeric_df.corr(),

cmap='coolwarm',

center=0

)

plt.title(
'Correlation Heatmap'
)

plt.show()
```
